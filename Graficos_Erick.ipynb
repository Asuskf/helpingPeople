```
import pandas as pd
import numpy as np
```

# Creación del set de datos

```
numero_elementos = 100
semilla = np.random.RandomState(numero_elementos)
cilindros = semilla.randint(1, 9, size=numero_elementos)
desplazamiento = semilla.randint(0, 40, size=numero_elementos)
Fuerza = semilla.randint(0, 24, size=numero_elementos)
peso = semilla.randint(0, 24, size=numero_elementos)
aceleracion = semilla.randint(0, 20, size=numero_elementos)
anio = semilla.randint(1,20, size=numero_elementos)
nombres = ['cilindro {}'.format(name) for name in semilla.randint(1, 4, size=numero_elementos)]
id_nombres = semilla.randint(1, 4, size=numero_elementos)
```

```
df = pd.DataFrame(
    {'cilindros': cilindros,
     'peso': peso,
     'aceleracion': aceleracion,
     'añio': anio,
     'nombres': nombres,
     'id_nombres': id_nombres
    })
```

# Coordenadas Paralelas

```
#Creación con Pandas
pd.plotting.parallel_coordinates(df, 'nombres', color=('#556270', '#4ECDC4', '#C7F464', '#C7F464'))
```

```
#Creación con la libreria Ploty
import plotly.express as px
df = df
fig = px.parallel_coordinates(df, color="id_nombres",
                              dimensions=['cilindros', 'peso', 'aceleracion', 'añio'],
                              color_continuous_scale=px.colors.diverging.Tealrose,
                              color_continuous_midpoint=2)
fig.show()
```

# Andrews curves

```
pd.plotting.andrews_curves(df, 'nombres')
```

# Gráfico Radial

```
import plotly.express as px
df = pd.DataFrame(dict(
    r=[1, 5, 2, 2, 3],
    theta=['processing cost','mechanical properties','chemical stability',
           'thermal stability', 'device integration']))

fig = px.line_polar(df, r='r', theta='theta', line_close=True)
fig.update_traces(fill='toself')
fig.show()
```

# Star plots

```
import plotly.graph_objects as go

categories = ['processing cost','mechanical properties','chemical stability',
              'thermal stability', 'device integration']

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=[1, 5, 2, 2, 3],
      theta=categories,
      fill='toself',
      name='Product A'
))
fig.add_trace(go.Scatterpolar(
      r=[4, 3, 2.5, 1, 2],
      theta=categories,
      fill='toself',
      name='Product B'
))

fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, 5]
    )),
  showlegend=False
)

fig.show()
```

# Contour plots

```
import plotly.graph_objects as go

fig = go.Figure(data =
    go.Contour(
        z=[[10, 10.625, 12.5, 15.625, 20],
           [5.625, 6.25, 8.125, 11.25, 15.625],
           [2.5, 3.125, 5., 8.125, 12.5],
           [0.625, 1.25, 3.125, 6.25, 10.625],
           [0, 0.625, 2.5, 5.625, 10]]
    ))
fig.show()
```
